**TASK 2**

```python
import re

def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)

    text = re.sub(r'\d+', '', text)

    text = text.lower()

    text = re.sub(r'(.)\1{2,}', r'\1', text)

    text = re.sub(r'[^a-z\s]', '', text)

    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    cleaned_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            cleaned_tokens.append(word)

    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence
```

**TASK 3**

```python
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10.",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

all_tokens = []

for text in sample_inputs:
    tokens, clean_sent = preprocess_text(text)
    all_tokens.extend(tokens)

    print("\nOriginal:", text)
    print("Tokens:", tokens)
    print("Cleaned:", clean_sent)
```

**TASK 4**

```python
def token_analytics(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total if total > 0 else 0
    return total, unique, avg_len

total, unique, avg = token_analytics(all_tokens)

print("\nTotal Tokens:", total)
print("Unique Tokens:", unique)
print("Average Token Length:", round(avg, 2))
```

**TASK 5**

```python
from collections import Counter

counter = Counter(all_tokens)

print("\nTop 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[:-6:-1])
```

**TASK 6**

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_sent = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_sent)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

result = full_pipeline(sample_inputs)

print(result)
```
